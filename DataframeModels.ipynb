# building 10 disease classes data for X-Y

```python
#get total numer of columns
df_cols=len(df.columns.tolist())
non_prevalence_cols=df_cols-11 #removing the date col as well
print(df_cols, non_prevalence_cols)
```

```python
#function for getting the database for prediction task
def get_train_test_xy(df, disease, split_perc=0.9):
    df=df.dropna()
    #df is the whole dataframe
    split_idx=int(split_perc*df.shape[0])
    trainX=df.values[:split_idx,3:non_prevalence_cols]
    trainY=df[disease].values[:split_idx]
    testX=df.values[split_idx:,3:non_prevalence_cols]
    testY=df[disease].values[split_idx:]
    return trainX, trainY, testX, testY

```

```python
disease_list=['chlamydia', 'gential_warts', 'gonorrhea', 'herpes', 'hpv',
              'other_std', 'parasitic', 'std_screen', 'syphilis', 'trich']

d='chlamydia'
trainX_ch, trainY_ch, testX_ch, testY_ch=get_train_test_xy(df, d)
print(trainX_ch.shape,trainY_ch.shape)
print(testX_ch.shape, testY_ch.shape)
print("-------------------test data-------------------")
print(testX_ch)
print(testY_ch)
```

```python
print(trainX_ch)
print(trainY_ch)
```

```python
from sklearn.svm import SVR
clf = SVR(C=1.0, epsilon=0.2)
#clf.fit(X, y)
clf.fit(trainX_ch, trainY_ch)
```

```python
pred=clf.predict(testX_ch)
acc=clf.score(testX_ch, testY_ch)

print("------------------test_predictions chlamydia-----------------")
print(pred)
print("------------------coeff R^2 pred chlamydia-------------------------")
print(acc)
```

```python
#Sklearn linear regression
from sklearn.linear_model import LinearRegression
reg=LinearRegression().fit(trainX_ch, trainY_ch)
score_train=reg.score(trainX_ch, trainY_ch)
score_test=reg.score(testX_ch, testY_ch)
print("--------------Linear Regression results--------------")
print("score_train=", score_train)
print("score_test", score_test)
```

```python
#CROSS VALIDATION------ Chlamydia
from sklearn import cross_validation, linear_model
trX_ch_npy=np.array(trainX_ch)
trY_ch_npy=np.array(trainY_ch)

loo = cross_validation.LeaveOneOut(len(trY_ch_npy))

regr = linear_model.LinearRegression()

scores = cross_validation.cross_val_score(regr, trX_ch_npy, trY_ch_npy, scoring='mean_squared_error', cv=loo,)

# This will print the mean of the list of errors that were output and 
# provide your metric for evaluation
print("-------------- Cross validation -----------------")
print("score")
print(scores.mean())
print("-------------------------------------------------")
```
